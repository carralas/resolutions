Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
orders
```

```python
customer_spent = orders[['CustomerID', 'amount_spent']].groupby('CustomerID').sum()
customer_spent
```

```python
#hard way
```

```python
customer_spent.loc[]
```

```python
customer_spent['amount_spent'].max()
```

```python
customer_spent['amount_spent'].min()
```

```python
range_spent = round(float(customer_spent.max()-customer_spent.min()), 2)
range_spent
```

```python
vipcut = round(range_spent*0.95, 2)
vipcut
```

```python
prefercut = round(range_spent*0.75, 2)
prefercut
```

```python
vipmask = customer_spent['amount_spent'] > vipcut
```

```python
prefermask = customer_spent['amount_spent'] > prefercut
```

```python
customer_spent[vipmask]
```

```python
customer_spent[~vipmask & prefermask]
```

```python
vipcut = len(customer_spent.index)*0.95
vipcut
```

```python
prefercut = len(customer_spent.index)*0.75
prefercut
```

```python
customer_spent.iloc[int(vipcut):, :]
```

```python
customer_spent.iloc[int(prefercut):int(vipcut), :]
```

```python
#easy way
```

```python
customer_spent.loc[customer_spent['amount_spent'] > customer_spent.quantile(q=0.75)[0], 'CustomerType'] = 'PRF'
customer_spent.loc[customer_spent['amount_spent'] > customer_spent.quantile(q=0.95)[0], 'CustomerType'] = 'VIP'
customer_spent.head(10)
```
