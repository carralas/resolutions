#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
states = list(temp['State'].unique())
```

```python
states
```

```python
for state in states:
    print(state)
    print('temperature avg: %.2f' % float(temp[temp['State'] == state].mean()))
    print('temperature std: %.2f' % float(temp[temp['State'] == state].std()))
```

#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
departments = list(employee['Department'].unique())
```

```python
for department in departments:
    print(department)
    print('years max: %.2f' % employee[employee['Department'] == department]['Years'].max())
    print('years mean: %.2f' % employee[employee['Department'] == department]['Years'].mean())
    print('yeas min: %.2f' % employee[employee['Department'] == department]['Years'].min())
    print('salary max: %.2f' % employee[employee['Department'] == department]['Salary'].max())
    print('salary mean: %.2f' % employee[employee['Department'] == department]['Salary'].mean())
    print('salary min: %.2f' % employee[employee['Department'] == department]['Salary'].min())
```

### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
for department in departments:
    print(f"{department} difference: {employee[employee['Department'] == department]['Salary'].max()-employee[employee['Department'] == department]['Salary'].min()}")
```
