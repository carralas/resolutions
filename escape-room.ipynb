```python
room_musical = {
    "name": "music room",
    "type": "room",
}
room_bedroom = {
    "name": "big bedroom",
    "type": "room",
}
room_kids = {
    "name": "kids bedroom",
    "type": "room",
}
room_dining = {
    "name": "dining room",
    "type": "room",
}
room_garden = {
    "name": "garden",
    "type": "room",
}
door_musical = {
    "name": "do re mi door",
    "type": "door",
    "locked": True
}
door_bedroom = {
    "name": "big door",
    "type": "door",
    "locked": True
}
door_kids = {
    "name": "colorful door",
    "type": "door",
    "locked": True
}
door_garden = {
    "name": "iron door",
    "type": "door",
    "locked": True
}
key_musical = {
    "name": "guitar-shaped key",
    "type": "key",
    "target": door_musical,
}
key_kids = {
    "name": "toy key",
    "type": "key",
    "target": door_kids,
}
key_bedroom = {
    "name": "golden key",
    "type": "key",
    "target": door_bedroom,
}
key_garden = {
    "name": "passcard",
    "type": "key",
    "target": door_garden,
}
furniture_couch = {
    "name": "old couch",
    "type": "furniture",
}
furniture_piano = {
    "name": "piano",
    "type": "furniture",
}
furniture_queenbed = {
    "name": "queen bed",
    "type": "furniture",
}
furniture_smallbed = {
    "name": "small bed",
    "type": "furniture",
}
furniture_dresser = {
    "name": "kids dresser",
    "type": "furniture",
}
furniture_table = {
    "name": "dining table",
    "type": "furniture",
}

rooms = [room_musical, room_bedroom, room_kids, room_dining, room_garden]
keys = [key_musical, key_bedroom, key_kids, key_garden]

object_relations = {
    "music room": [door_musical, furniture_couch, furniture_piano],
    "big bedroom": [door_musical, door_kids, door_bedroom, furniture_queenbed],
    "kids bedroom": [door_kids, furniture_smallbed, furniture_dresser],
    "dining room": [door_bedroom, door_garden, furniture_table],
    "garden": [door_garden],
    "do re mi door": [room_musical, room_bedroom],
    "big door": [room_bedroom, room_dining],
    "colorful door": [room_bedroom, room_kids],
    "iron door": [room_dining, room_garden],
    "piano": [key_musical],
    "queen bed": [key_kids],
    "small bed": [key_bedroom],
    "kids dresser": [key_garden],
}

def playroom(room):
    gamestate["currentroom"] = room
    if gamestate["currentroom"] == gamestate["target"]: print("You got out! But it's raining... Let's go back inside.")
    else:
        print("You are currently at " + room["name"] + ".")
        action = input("What to do now? (explore/examine) ").strip()
        if action == "explore":
            exploreroom(room)
            playroom(room)
        elif action == "examine": examineitem(input("What wold you like to examine? ").strip())
        else:
            print("invalid action.")
            playroom(room)

def exploreroom(room):
    furniture = [k["name"] for k in object_relations[room["name"]]] #pega o nome do quarto, checa as relações e lista
    print("In " + room["name"] + " you find " + ", ".join(furniture) + ".")

def examineitem(furniture):
    currentroom = gamestate["currentroom"] #designa o quarto atual descrito no gamestate
    nextroom = ''
    output = 'nothing'
    for item in object_relations[currentroom["name"]]: #pega o nome do quarto e checa as relações
        if item["name"] == furniture:
            output = "You examine " + furniture + "."
            if item["type"] == "door":
                havekey = False
                for key in gamestate["bag"]:
                    if key["target"] == item: havekey = True
                if (havekey):
                    if item["locked"] == True:
                        item["locked"] = False
                        output += " You unlocked the door."
                        nextroom = crossdoor(item, currentroom)
                    else:
                        output += " It's unlocked."
                        nextroom = crossdoor(item, currentroom)
                    output += " You can see " + nextroom["name"] + " from here."
                else: print(" It's locked.")
            else:
                if item["name"] in object_relations:
                    founditem = object_relations[item["name"]].pop()
                    output += " You find " + founditem["name"] + "."
                    gamestate["bag"].append(founditem) #adiciona a chave à mochila e deleta do item removendo ela da lista
                else:
                    output = "Nothing here."
            print(output)
            break #garante que só roda uma vez
    if output == 'nothing': print("This object does not exist.")
    if nextroom and input("Do you want to go to next room? (yes/no) ").strip() == 'yes': playroom(nextroom)
    else: playroom(currentroom)

def crossdoor(door, currentroom):
    for room in object_relations[door["name"]]:
        if currentroom != room: return room

gamestate = {
    "currentroom": room_musical,
    "bag": [],
    "target":  room_garden,
}

print("It's carnival and you don't remember where you left your keys after last night's street party. Let's look for it!\n")
playroom(gamestate["currentroom"])
```
