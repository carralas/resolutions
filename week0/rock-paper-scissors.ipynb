```python
import random

status = ['stone', 'paper', 'scissors']
games = 0
playerscore = 0
machinescore = 0
maxgames = 5
endcondition = maxgames/2+1

def choice():
    pick = 'nothing'
    while pick != 'stone' and pick != 'paper' and pick != 'scissors':
        pick = input("Choose from stone, paper or scissors: ")
        if pick != 'stone' and pick != 'paper' and pick != 'scissors': print("Not a valid choice.")
    return pick
            
def combat():
    if player == machine: return 0
    else:
        if player == 'stone':
            if machine == 'paper': return 1
            else: return 2
        elif player == 'paper':
            if machine == 'stone': return 2
            else: return 1
        else:
            if machine == 'stone': return 1
            else: return 2
                

while games < maxgames and playerscore < endcondition and machinescore < endcondition:
    machine = random.choice(status)
    player = choice()
    result = combat()
    print("You choose", player, "\nThe machine choose", machine)
    if result == 0:
        print("Tie!\n")
    elif result == 1:
        print("You lose!\n")
        machinescore += 1
    else:
        print("You win!\n")
        playerscore += 1
    games += 1
print("\nTotal score\nYou:", playerscore, "- Machine:", machinescore)
if playerscore == machinescore: print("It's a tie!")
elif playerscore < machinescore: print("You lost!")
else: print("You won!")
```

```python
import random

status = ['stone', 'paper', 'scissors', 'lizard', 'spock']
games = 0
playerscore = 0
machinescore = 0
maxgames = 0
endcondition = 0
endcondition = maxgames/2+1

def choice():
    pick = 'nothing'
    while pick != 'stone' and pick != 'paper' and pick != 'scissors' and pick != 'lizard' and pick != 'spock':
        pick = input("Choose from stone, paper, scissors, lizard or spock: ")
        if pick != 'stone' and pick != 'paper' and pick != 'scissors' and pick != 'lizard' and pick != 'spock': print("Not a valid choice.")
    return pick
            
def combat():
    if player == machine: return 0
    else:
        if player == 'stone':
            if machine == 'paper' or machine == 'spock': return 1
            else: return 2
        elif player == 'paper':
            if machine == 'stone' or machine == 'spock': return 2
            else: return 1
        elif player == 'scissors':
            if machine == 'stone' or machine == 'spock': return 1
            else: return 2
        elif player == 'lizard':
            if machine == 'stone' or machine == 'scissors': return 1
            else: return 2
        else:
            if machine == 'stone' or machine == 'scissors': return 2
            else: return 1
                
while maxgames%2 != 1:
    maxgames = int(input("How many matches are you playing? "))
    if maxgames%2 != 1: print("Please select an odd number os matches.")
endcondition = int(maxgames/2+1)
print("The winner must win", endcondition, "matches.\n")
while games < maxgames and playerscore < endcondition and machinescore < endcondition:
    machine = random.choice(status)
    player = choice()
    result = combat()
    print("You choose", player, "\nThe machine choose", machine)
    if result == 0:
        print("Tie!\n")
    elif result == 1:
        print("You lose!\n")
        machinescore += 1
    else:
        print("You win!\n")
        playerscore += 1
    games += 1
print("\nTotal score\nYou:", playerscore, "- Machine:", machinescore)
if playerscore == machinescore: print("It's a tie!")
elif playerscore < machinescore: print("You lost!")
else: print("You won!")
```
