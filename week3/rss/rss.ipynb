### 1. Use feedparser to parse the following RSS feed URL.

### 4. Extract and print the feed title, subtitle, author, and link.

```python
to_get = ['title', 'subtitle', 'feedburner_feedburnerhostname', 'link']
infos = []

for item in to_get:
    infos.append(eval(f'feed.feed.{item}'))
infos
```

### 7. Extract a list of entry titles.

```python
titles = [feed.entries[k].title for k in range(len(feed.entries))]
titles
```

### 8. Calculate the percentage of "Four short links" entry titles.

```python
k = 0

for item in titles:
    if 'Four short links' in item: k += 1
k/len(titles)
```

### 9. Create a Pandas data frame from the feed's entries.

```python
import pandas as pd
```

```python
entriesdf = pd.DataFrame(feed.entries)
entriesdf
```

### 10. Count the number of entries per author and sort them in descending order.

```python
countentriesdf = entriesdf.groupby('author', as_index=False).agg({'title':'count'})
countentriesdf.columns = ['author', 'entries']
countentriesdf
```

### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
entriesdf['title_length'] = entriesdf['title'].apply(len)
entriesdf[['author', 'title', 'title_length']].sort_values('title_length', ascending=False)
```

### 12. Create a list of entry titles whose summary includes the phrase "machine learning."

```python
machinelist = []

summaries = [feed.entries[k].summary for k in range(len(feed.entries))]
for item in summaries:
    if 'machine learning' in item.lower(): machinelist.append(item)
machinelist
```
