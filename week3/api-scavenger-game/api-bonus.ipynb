```python
coins_of_interest = ['bitcoin', 'dogecoin']
coins_id = []

for dic in coins_info_returns:
    if dic['name'].lower() in coins_of_interest: coins_id.append({'name': dic['name'], 'id': dic['id']})
```

```python
import datetime
```

```python
lastmonth = []

for k in range(30): lastmonth.append(str(datetime.date.today()+datetime.timedelta(-k)))
```

```python
lastmonth
```

```python
for k in range(30):
    lastmonth[k] = f'{lastmonth[k][-2:]}-{lastmonth[k][5:7]}-{lastmonth[k][:4]}'
```

```python
lastmonth
```

```python
dogecoin_history = []
bitcoin_history = []
for date in lastmonth:
    dogecoin_history.append(requests.get(f'{baseurl}coins/dogecoin/history?date={date}'))
    bitcoin_history.append(requests.get(f'{baseurl}coins/bitcoin/history?date={date}'))
```

```python
dogecoin_history
```

```python
bitcoin_history
```

```python
dogecoin_history_returns = [response.json() for response in dogecoin_history]
bitcoin_history_returns = [response.json() for response in bitcoin_history]
```

```python
dogecoin_prices = []
bitcoin_prices = []
for k in range(30):
    dogecoin_prices.append(dogecoin_history_returns[k]['market_data']['current_price']['usd'])
    bitcoin_prices.append(bitcoin_history_returns[k]['market_data']['current_price']['usd'])
```
